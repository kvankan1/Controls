```python
m = 4
b = 9
k = 60
F = 1

A = np.matrix([[-b/m, -k/m], [1, 0]])
B = np.matrix([[F/m], [0]])
C = np.matrix([[0, 1]])
D = np.matrix([[0]])

sys = ss(A,B,C,D)
print(sys)

x0 = np.matrix([[1],[0]])

t = np.arange(0, 20, 0.01)

y1, t = initial(sys, t, x0)

plt.plot(t, y1)
plt.vlines([1.693], -0.15, 0.15, colors ="red", label = "Period")
plt.legend()
plt.show()
```
