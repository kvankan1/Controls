Transfer Functions

```python
#Create Laplace variable 's'
s = c.tf([1,0], [1])
#Note, this is actually just specifying the order & coefficients of the numerators, 
#and then the order & coefficents of the denominators

#and then create a transfer function with 's'
h = 1/(2*s + 1)
```

Combining Transfer Functions

```python
#Create 2 Transfer Functions
s = c.tf([1,0], [1])

h1 = 1/((1+2*s)*(1+3*s))
h2 = 1/(1+3*s)

h_sum = h1 + h2
print("h1+h2 =", h_sum)
```

```python
#Use minreal after operations to simply the expressions
h_sum = (h1+h2).minreal()

print("h1+h2 (simplified) =", h_sum)
```

Implementing Feedback



```python
#Use function feedforward.feedback(feedback)

h_cl = h1.feedback(h2)
print(h_cl)
```
